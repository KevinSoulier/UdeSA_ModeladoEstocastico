### Simulación de datos

En Modelado Estocástico trabajaremos continuamente con variables aleatorias.

````python
np.random.seed(123)

muestra = np.random.normal(
    loc=0,
    scale=1,
    size=10
)

muestra
````

### DataFrames con Pandas

Gran parte de los datos económicos y financieros se almacenan, procesan y/o visualizan en tablas. Para el consumo de datos, su procesamiento y análisis con frecuencia vamos a utilizar [Pandas](https://pandas.pydata.org/pandas-docs/stable/getting_started/intro_tutorials/index.html). Esta biblioteca nos provee muchas herramientas para manipular datos estructurados en dataframes, además es compatible con otros recursos útiles.

````python
df = pd.DataFrame({
    "Fecha":[1,2,3,4,5],
    "Precio":[100,105,103,110,108]
})

df
````

````python
df["Precio"]
````

````python
df["Precio"].mean()
````

### Gráficos

Para graficar vamos a utilizar `pyplot` de `matplotlib`.

Por ejemplo, para hacer un gráfico de líneas:

````python
plt.figure(figsize=(8,4))

plt.plot(df["Fecha"],
         df["Precio"])

plt.xlabel("Tiempo")
plt.ylabel("Precio")
plt.title("Serie temporal")
plt.show()
````

Para generar un histograma de los datos en la variable `muestra` creada anteriormente ejecutamos:

````python
plt.figure(figsize=(8,4))

plt.hist(
    muestra,
    bins=10
)

plt.title("Histograma")
plt.show()
````
